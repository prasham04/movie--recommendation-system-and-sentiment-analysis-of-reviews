# 2018 #

```python

import pandas as pd
import numpy as np
```

```python

link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = pd.concat([df1,df2,df3,df4],ignore_index=True)
```

```python
import json
import requests
from typing import Dict,Any
OMDB_API_KEY = '30dbbe0d'
```

```python
def get_movie_info(title: str) -> Dict[str, Any]:
    url = f"http://www.omdbapi.com/?t={title}&apikey={OMDB_API_KEY}"
    response = requests.get(url)
    data = response.json()
    
    if data['Response'] == 'True':
        return {
            'director_name': data.get('Director', 'unknown'),
            'actor_1_name': data.get('Actors', '').split(', ')[0] if data.get('Actors') else 'unknown',
            'actor_2_name': data.get('Actors', '').split(', ')[1] if len(data.get('Actors', '').split(', ')) > 1 else 'unknown',
            'actor_3_name': data.get('Actors', '').split(', ')[2] if len(data.get('Actors', '').split(', ')) > 2 else 'unknown',
            'genres': data.get('Genre', 'unknown'),
            'movie_title': data.get('Title', 'unknown').lower()
        }
    else:
        return {
            'director_name': 'unknown',
            'actor_1_name': 'unknown',
            'actor_2_name': 'unknown',
            'actor_3_name': 'unknown',
            'genres': 'unknown',
            'movie_title': title.lower()
        }
```

```python
new_df18 = pd.DataFrame(df['Title'].apply(get_movie_info).tolist())
```

```python
new_df18['director_name']=new_df18['director_name'].apply(lambda x: x.replace(',',' '))
```

```python
new_df18['genres']=new_df18['genres'].apply(lambda x: x.replace(',',' '))
```

```python
new_df18['comb'] = (new_df18['actor_1_name'] + ' ' + 
                    new_df18['actor_2_name'] + ' ' + 
                    new_df18['actor_3_name'] + ' ' + 
                    new_df18['director_name'] + ' ' + 
                    new_df18['genres'])
```

```python
new_df18
```

```python
unknown_counts =((new_df18 == 'unknown') | new_df18.isna()).sum()
unknown_counts
```

```python
df_18_cleaned = new_df18[~(new_df18 == 'unknown').any(axis=1)]
df_18_cleaned
```

```python
unknown_counts =((df_18_cleaned == 'unknown') | df_18_cleaned.isna()).sum()
unknown_counts
```
